# Temel Bulgular:

Küresel Geri Dönüşüm: Çoğu ülkenin düşük-orta geri dönüşüm oranlarına sahip olduğu.
En Büyük Atık Üreticileri: Çin ve ABD'nin öne çıktığı, Türkiye'nin bu listede olmadığı.
Ana Atık Kaynağı: Tüketici ambalajlarının yaygınlığı, Türkiye'nin de bu eğilimde olduğu.
Kişi Başı Atık vs. Geri Dönüşüm: Belirgin bir doğrusal ilişki olmadığı, Türkiye'nin düşük geri dönüşümle orta düzeyde kişi başı atığa sahip olduğu.
Kıyı Atık Riski: Yüksek riskli ülkelerde genellikle düşük geri dönüşüm oranları, Türkiye'nin de bu kategoride yer aldığı.
Türkiye Özeti: Orta düzeyde atık üretimi, düşük geri dönüşüm oranı, yüksek kıyı riski ve ana kaynak olarak tüketici ambalajları.

En Çok Plastik Atık Üreten 10 Ülke

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dosya = 'Plastic Waste Around the World.csv'
df = pd.read_csv(dosya)

top10_waste = df.sort_values(by="Total_Plastic_Waste_MT", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_waste, x="Total_Plastic_Waste_MT", y="Country", palette="Reds_r")
plt.title(" En Çok Plastik Atık Üreten 10 Ülke")
plt.xlabel("Toplam Plastik Atık (Milyon Ton)")
plt.ylabel("Ülke")
plt.tight_layout()
plt.show()
```

Kıyı atık riskine göre gruplandırma ve ortalama geri dönüşüm oranlarını hesaplama/ Pasta Grafiği


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dosya = 'Plastic Waste Around the World.csv'
df = pd.read_csv(dosya)
average_recycling_by_risk = df.groupby('Coastal_Waste_Risk')['Recycling_Rate'].mean().sort_values(ascending=False)
print("\nOrtalama Geri Dönüşüm Oranları (Kıyı Atık Riskine Göre):")
print(average_recycling_by_risk)

plt.figure(figsize=(8, 8))
plt.pie(average_recycling_by_risk, labels=average_recycling_by_risk.index, autopct='%1.1f%%', startangle=140)
plt.title('Kıyı Atık Riskine Göre Ortalama Geri Dönüşüm Oranlarının Dağılımı')
plt.ylabel('Ortalama Geri Dönüşüm Oranı (%)') # Y ekseni etiketi eklendi
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Geri Dönüşüm Oranı ve Kıyı Atık Riski Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dosya = 'Plastic Waste Around the World.csv'
df = pd.read_csv(dosya)

 
yuksek_risk_df = df[df['Coastal_Waste_Risk'] == 'High']
dusuk_risk_df = df[df['Coastal_Waste_Risk'] == 'Low']
# Box plot oluşturma
plt.figure(figsize=(8, 6))
sns.boxplot(x='Coastal_Waste_Risk', y='Recycling_Rate', data=pd.concat([yuksek_risk_df, dusuk_risk_df]),
            order=['High', 'Low']) 
plt.title(f'{'High'} ve {'Low'} Kıyı Atık Riskine Sahip Ülkelerin Geri Dönüşüm Oranları')
plt.xlabel('Kıyı Atık Riski')
plt.ylabel('Geri Dönüşüm Oranı (%)')
plt.show()
```

Toplam Atığı En Yüksek Ve Düşük 10 Ülkedeki Ana Atık Kaynağı Sıklıklarının Karşıllaştrılması

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dosya= 'Plastic Waste Around the World.csv'
df = pd.read_csv(dosya)

df_sorted_by_total_desc = df.sort_values(by='Total_Plastic_Waste_MT', ascending=False)
df_sorted_by_total_asc = df.sort_values(by='Total_Plastic_Waste_MT', ascending=True)


top_10_waste = df_sorted_by_total_desc.head(10)
bottom_10_waste = df_sorted_by_total_asc.head(10)


top_10_waste['Grup'] = 'En Yüksek Atık (İlk 10)'
bottom_10_waste['Grup'] = 'En Düşük Atık (İlk 10)'

combined_waste_groups = pd.concat([top_10_waste, bottom_10_waste])


plt.figure(figsize=(12, 6))
sns.countplot(data=combined_waste_groups, x='Grup', hue='Main_Sources')
plt.title('En Yüksek ve En Düşük Atığa Sahip 10 Ülkenin Ana Kaynak Dağılımı')
plt.xlabel('Atık Grubu')
plt.ylabel('Ülke Sayısı')
plt.xticks(rotation=0)
plt.legend(title='Ana Kaynaklar')
plt.tight_layout()
plt.show()
```

Türkiye'nin Plastik Atık Üretimi, Geri Dönüşümündeki Yeri

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dosya = 'Plastic Waste Around the World.csv'
df = pd.read_csv(dosya)
turkiye_verisi = df[df['Country'] == 'Turkey']

ulke = turkiye_verisi['Country'].iloc[0]
toplam_atik = turkiye_verisi['Total_Plastic_Waste_MT'].iloc[0]
kaynaklar = turkiye_verisi['Main_Sources'].iloc[0]
geri_donusum_orani = turkiye_verisi['Recycling_Rate'].iloc[0]
kisi_basi_atik = turkiye_verisi['Per_Capita_Waste_KG'].iloc[0]
kiyi_riski = turkiye_verisi['Coastal_Waste_Risk'].iloc[0]

print("Türkiye'nin Verileri:")
print(f"Ülke: {ulke}")
print(f"Toplam Plastik Atık (Milyon Ton): {toplam_atik}")
print(f"Ana Kaynaklar: {kaynaklar}")
print(f"Geri Dönüşüm Oranı (%): {geri_donusum_orani}")
print(f"Kişi Başına Düşen Atık (KG): {kisi_basi_atik}")
print(f"Kıyı Atık Riski: {kiyi_riski}")
```
